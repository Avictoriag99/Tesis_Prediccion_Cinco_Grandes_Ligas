Esto redujo el conjunto inicial de variables a 8. Posteriormente, se aplicó Lasso, que, mediante una técnica de regresión con penalización, reduce los coeficientes de las variables menos influyentes a cero, eliminándolas del modelo. ##### Se aplicara validación cruzada, con el modelo de variables por encima de 0.6 en la correlación, toda vez que fue el modelo que mejores valores alcanzo

```python
from sklearn.model_selection import cross_validate, KFold
from sklearn.linear_model import LinearRegression
import numpy as np
## SE APLICO LA VALIDACIÓN CRUZADA CON 5 FOLDS

# Preparar los datos para el modelo
X = df_defensores[variables_significativas_originales1]
y = df_defensores['Precio 2023-2024']

# Configurar la validación cruzada
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Crear el modelo de regresión lineal
model = LinearRegression()

# Realizar la validación cruzada con múltiples métricas
results = cross_validate(model, X, y, cv=kf, scoring=['neg_mean_squared_error', 'r2'],
                         return_train_score=False)

# Calcular el MSE (los valores son negativos, así que los multiplicamos por -1)
mse_scores = -results['test_neg_mean_squared_error']

# Calcular el RMSE a partir del MSE negativo
rmse_scores = np.sqrt(mse_scores)

# Configurar numpy para que no use notación científica
np.set_printoptions(suppress=True)

# Imprimir los resultados en formato "normal"
print("R-squared (R2) para cada fold:")
for r2 in results['test_r2']:
    print(f"{r2:.6f}")  # Imprime con 6 decimales

print(f"R2 promedio: {np.mean(results['test_r2']):.6f}")

print("\nMean Squared Error (MSE) para cada fold:")
for mse in mse_scores:
    print(f"{mse:,.2f}")  # Imprime con 2 decimales y comas como separadores de miles

print(f"MSE promedio: {np.mean(mse_scores):,.2f}")

print("\nRoot Mean Squared Error (RMSE) para cada fold:")
for rmse in rmse_scores:
    print(f"{rmse:,.2f}")  # Imprime con 2 decimales y comas como separadores de miles

print(f"RMSE promedio: {np.mean(rmse_scores):,.2f}")
```

##### Se aplicara validación cruzada, con el modelo de variables por encima de 0.7 en la correlación, toda vez que fue el modelo que mejores valores alcanzo

```python
## SE APLICO LA VALIDACIÓN CRUZADA CON 5 FOLDS

from sklearn.model_selection import cross_validate, KFold
from sklearn.linear_model import LinearRegression
import numpy as np

# Preparar los datos para el modelo
X = mc_mediocampistas[variables_significativas_mc2]
y = mc_mediocampistas['Precio 2023-2024']

# Configurar la validación cruzada
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Crear el modelo de regresión lineal
model = LinearRegression()

# Realizar la validación cruzada con múltiples métricas
results = cross_validate(model, X, y, cv=kf, scoring=['neg_mean_squared_error', 'r2'],
                         return_train_score=False)

# Calcular el MSE (los valores son negativos, así que los multiplicamos por -1)
mse_scores = -results['test_neg_mean_squared_error']

# Calcular el RMSE a partir del MSE negativo
rmse_scores = np.sqrt(mse_scores)

# Configurar numpy para que no use notación científica
np.set_printoptions(suppress=True)

# Imprimir los resultados en formato legible
print("R-squared (R2) para cada fold:")
for r2 in results['test_r2']:
    print(f"{r2:.6f}")  # Imprime con 6 decimales

print(f"R2 promedio: {np.mean(results['test_r2']):.6f}")

print("\nMean Squared Error (MSE) para cada fold:")
for mse in mse_scores:
    print(f"{mse:,.2f}")  # Imprime con 2 decimales y comas como separadores de miles

print(f"MSE promedio: {np.mean(mse_scores):,.2f}")

print("\nRoot Mean Squared Error (RMSE) para cada fold:")
for rmse in rmse_scores:
    print(f"{rmse:,.2f}")  # Imprime con 2 decimales y comas como separadores de miles

print(f"RMSE promedio: {np.mean(rmse_scores):,.2f}")
```
